**Lesson 1: Variables & Printing 🐍**
What is a variable?
Think of a variable as a labelled box. You put something inside it and give it a name. Later you can open it by calling its name.

```python
name = "Rahul"
age  = 25
city = "Pune"
```

The = sign means "store this value in this box". 

## 📌 Data Types in Python

There are **3 basic types of values** to understand:

|        Type          |     Example   |        Rule            |
|        ------        |     --------  |       ------           |
| Text (string)        | `"Hello"`     | Always use quotes      |
| Number               | `25`          | No quotes              |
| Boolean (True/False) | `True`        | Use capital `T` or `F` |



**Printing to the screen**

print() is how Python speaks. Whatever you put inside the brackets gets shown.

```python
print("Hello, World!")
print(42)
```

You can also print a variable by just using its name:

```python
name = "Rahul"
print(name)
```

**Mixing text + variables (f-strings)**
This is where it gets fun. Put f before the quote, then wrap your variable in {curly braces}:

```python
name = "Rahul"
age  = 25
city = "Pune"

print(f"My name is {name}")
print(f"I am {age} years old")
print(f"I live in {city}")
```

Python even does math inside the curly braces:

```python
price    = 100
discount = 20
print(f"You pay: {price - discount} rupees")
```
